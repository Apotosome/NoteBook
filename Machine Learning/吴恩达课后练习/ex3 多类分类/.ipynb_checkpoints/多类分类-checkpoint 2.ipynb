# 多类别分类 Multi-class Classification 

在本部分的练习中，您将使用逻辑回归和神经网络识别手写数字（从0到9）。同时我们将拓展ex2中的逻辑回归的实现。

实现原理可以参考`week3 6.7 多类别分类：一对多`

为了能实现这样的转变，我们将多个类中的一个类标记为正向类（$y=1$），然后将其他所有类都标记为负向类，这个模型记作$h_\theta^{\left( 1 \right)}\left( x \right)$。接着，类似地第我们选择另一个类标记为正向类（$y=2$），再将其它类都标记为负向类，将这个模型记作 $h_\theta^{\left( 2 \right)}\left( x \right)$,依此类推。

我们得到一系列的模型简记为： 
$$
h_\theta^{\left( i \right)}\left( x \right)=p\left( y=i|x;\theta  \right)
$$
​	其中：$i=\left( 1,2,3....k \right)$ 

在我们需要做预测时，我们将所有的分类机都运行一遍，然后对每一个输入变量，都选择最高可能性的输出变量。

即选择一个让$h_\theta^{\left( i \right)}\left( x \right)$ 最大的$ i$
$$
\mathop{\max}\limits_i\,h_\theta^{\left( i \right)}\left( x \right)
$$


选择出哪一个分类器是可信度最高效果最好的，那么就可认为得到一个正确的分类

## 正则化代价函数 regularized cost

我们可以将逻辑回归实现修改为完全向量化。这是因为向量化代码除了简洁外，还能够利用线性代数优化，并且通常比迭代代码快得多。

正则化的逻辑回归模型
$$
J\left( \theta  \right)=\frac{1}{m}\sum\limits_{i=1}^{m}{[-{{y}^{(i)}}\log \left( {h_\theta}\left( {{x}^{(i)}} \right) \right)-\left( 1-{{y}^{(i)}} \right)\log \left( 1-{h_\theta}\left( {{x}^{(i)}} \right) \right)]}+\frac{\lambda }{2m}\sum\limits_{j=1}^{n}{\theta _{j}^{2}}
$$

其中：
$$
h_\theta\left( x \right)=g(\theta^TX)\\
g\left( z \right)=\frac{1}{1+{{e}^{-z}}}\\
$$

```python
def sigmoid(z):
    return 1 / (1+np.exp(-z))
```

```python
def computerCostReg(theta,X,y,Rate=1):
    X = np.matrix(X)   # 5000 * 400
    y = np.matrix(y)   # 5000 * 1
    theta = np.matrix(theta)  # 400 * 1 
    
    parameters = theta.shape[1]
    
    first = np.multiply(-y,np.log(sigmoid(X * theta.T)))
    second = np.multiply((1-y),np.log(1-sigmoid(X * theta.T)))
    reg = (Rate/(2*len(X))) * np.sum(np.power(theta[:,1:parameters],2))
    
    return np.sum(first - second) / len(X) + reg
```

## 正则化梯度 regularized gradient

  如果我们要**使用梯度下降法**令这个代价函数最小化，因为我们未对$\theta_0$进行正则化，所以梯度下降算法将分两种情形：
$$
\begin{array}{l}
\text { Repeat \{  } \\
\qquad 
{\theta_0}:={\theta_0}-a[\frac{1}{m}\sum\limits_{i=1}^{m}{(({h_\theta}({{x}^{(i)}})-{{y}^{(i)}})x_{0}^{(i)}})]
\\
\qquad 
{\theta_j}:={\theta_j}-a[\frac{1}{m}\sum\limits_{i=1}^{m}{(({h_\theta}({{x}^{(i)}})-{{y}^{(i)}})x_{j}^{\left( i \right)}}+\frac{\lambda }{m}{\theta_j}]
\\
\qquad(j = 1,2,3,...,n)  \\
\text { \} }
\end{array}
$$



在此我们定义梯度函数即可

$$
\frac{\partial J(\theta)}{\partial \theta_{0}}=\frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x^{(i)})-y^{(i)}) x_{j}^{(i)} \quad \text { for } j = 0 \\
\frac{\partial J(\theta)}{\partial \theta_{j}}=\left(\frac{1}{m} \sum_{i=1}^{m}\left(h_{\theta}\left(x^{(i)}\right)-y^{(i)}\right) x_{j}^{(i)}\right)+\frac{\lambda}{m} \theta_{j} \quad \text { for } j \geq 1
$$

将上式向量化可得：

$$
\frac{\partial J(\theta)}{\partial \theta_{0}}=\frac{1}{m} \sum_{i=1}^{m}(h_{\theta}(x^{(i)})-y^{(i)}) x_{j}^{(i)} \quad \text { for } j = 0 \\
\frac{\partial J(\theta)}{\partial \theta_{j}}=\frac{1}{m} X^T(h_{\theta}\left(x\right)-y) +\frac{\lambda}{m} \theta_{j} \quad \text { for } j \geq 1
$$

```python
def gradientReg(theta,X,y,Rate=1):
    X = np.matrix(X)
    y = np.matrix(y)
    theta = np.matrix(theta)
    
    parameters = theta.shape[1]
    error = sigmoid(X * theta.T) - y
    
    # (400*5000) * (5000*1) =  (400*1) 所求的矩阵大小 
    # 需要转置成(1*400)才能相加
    grad = (X.T * error / len(X)).T + ((Rate / len(X)) * theta)
    
    grad[0,0] = np.sum(np.multiply(error,X[:,0])) / len(X)  # for j=0
    
    grad = np.array(grad).ravel()  # 将(400*1)的矩阵展平
    return grad
```

## 训练模型

接下来我们将构建分类器。对于这个任务，我们有10个可能的类，并且由于逻辑回归只能一次在2个类之间进行分类，我们需要多类分类的策略。

在本练习中，我们的任务是实现一对一全分类方法，其中具有k个不同类的标签就有k个分类器，每个分类器在“是”和“不是”之间决定。 我们将把分类器训练包含在一个函数中，该函数计算10个分类器中的每个分类器的最终权重，并将权重返回为数组。

### 训练多维模型

```python
# 检查标签类型
np.unique(y)
```

```python
rows = X.shape[0]     # 5000 -- 5000个训练示例

# 添加新的一列 作为x0的系数  
# np.ones()返回一个全1的n维数组 axis=1 按照列方向
X = np.insert(X, 0, values=np.ones(rows), axis=1)    # (5000, 401)

cols = X.shape[1]     # 401
```

```python
import scipy.optimize as opt
def one_vs_all(X,y,number,Rate): # number分类器个数
    # （10,401）的零矩阵 用于存储多维模型每个分类器的θ参数的大θ
    all_theta = np.zeros((number,cols))
    
    # 每一次扫描一行存入all_theta 即 训练一个分类器
    for i in range(1,number+1):   
        theta_i = np.zeros(cols)  # (401,1)作为初值的小θ
        
        # 记录与正向类像匹配的标签并存入数组 
        # 就可以实现转换为逻辑回归问题 每个分类器在“1”和“0”之间决定。
        y_i = np.array([1 if label==i else 0 for label in y])
        # 将数组竖置   (5000, 1)
        y_i = np.reshape(y_i,(rows,1))
        
        #利用minimize实现梯度下降 找到最小
        result = opt.minimize(fun=computerCostReg,
                              x0=theta_i,
                              args=(X,y_i),
                              method='TNC',
                              jac = gradientReg
                             )
        all_theta[i-1,:] = result.x   # (1,401)
    return all_theta
        
```

```python
all_theta = one_vs_all(X,y,10,1)   # (10, 401)
```

### 用训练集预测和验证

逻辑回归模型的假设
$$
h_\theta \left( x \right)=g\left(\theta^{T}X \right)=\frac{1}{1+{{e}^{-\theta^{T}X}}}
$$

我们选择训练集中$h(x)$的最大值作为预测结果

```python
def predict(X,all_theta):
    X = np.matrix(X)
    all_theta = np.matrix(all_theta)
    
    h = sigmoid(X * all_theta.T)
    h_max = np.argmax(h,axis=1)  # 按行取 返回索引 由于数组从0开始因此+1
    return h_max + 1
```

```python
prediction = predict(X,all_theta)
prediction[3500:3510].ravel()
```

```python
y[3500:3510].ravel()
```

```python
# 对比目标变量y 计算精准度
# zip() 函数用于将可迭代的对象作为参数，将对象中对应的元素打包成一个个元组，然后返回由这些元组组成的列表。
def accuracy(predictions,y):
    currectList = []
    for (a,b) in zip(predictions,y):
        if a==b :
            currectList.append(1)
        else:
            currectList.append(0)
    accuracy = sum(map(int,currectList)) / len(currectList)  # 精准度
    return accuracy,currectList
```

```python
acc,acc_list = accuracy(predictions=prediction,y=y)
acc_list[:10]
```

```python
print('TNC--精准度为{}%'.format(acc*100))
```

```python
from sklearn.metrics import classification_report   #这个包是评价报告
print(classification_report(y,prediction))
```
